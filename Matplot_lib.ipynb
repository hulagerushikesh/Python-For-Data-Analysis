```python
#                                       DATA VISUALIZATION :
# -----> In simple , Is a graphical representation of the data in order to make a convenient insights of bunch of data 
# even dense data.  By using visual elements like charts, graphs, and maps, data visualization tools provide an 
# accessible way to see and understand trends, outliers, and patterns in data.

# Lets explore the python for data visualization.

#                                   @Matplotlib
# ---->  Is a 2D and 3D plotting python library.
# ----> Need of matplotlib is to plot dense data in order to understand the behaviour of data.


# 1) Line plot : Line Plots display numerical values on one axis, and categorical values on the other.
import numpy as np 
import matplotlib.pyplot as plt
from matplotlib import style
style.use('ggplot')
plt.figure(figsize=(16,12))
days = np.arange(1 , 31)
temp = np.linspace(25 , 40 , 30)
temp1 = np.array([23,34,45,23,22,33,43,44,34,25,27,32,29,39,31, 
                  23,34,46,42,39,31,27,23,26,29,33,37,31,30,40])
plt.plot(days , temp , 'go--' , linewidth = 2 , markersize=10 , label='temp line' )    # plt.plot(x , y , 'go--')
plt.plot(days , temp1 , 'yo:' , linewidth = 2 , markersize=10 , label='temp1 line' ) 
#plt.axis([0 , 20 , 0 , 50])   #plt.axis([xmin , xmax , ymin , ymax])
plt.title('Temperature' , fontsize = 20)
plt.xlabel('Days' , fontsize = 15)
plt.ylabel('Temp' , fontsize = 15)
plt.legend( loc = 0)
plt.grid(color ='r' , linestyle = '-' , linewidth = 1)
plt.show()
```

```python
# 2)Histagram : A histogram shows the frequency on the vertical axis and the horizontal axis is another dimension.
import random
import numpy as np
import matplotlib.pyplot as plt


a = np.random.randint(18,30,(50))   # np.random.randint(start , end , (total))
b = np.random.randint(20,40,(50)) 
print(a)
print(b)
plt.hist([a ,b], rwidth=0.9 , color = ['b' , 'r'] , orientation = 'vertical') # by chang ing orientation to horzontal 
                                                                              # we can get the horizontal histagram
#plt.hist(b , rwidth=0.9 , color = 'r')
plt.title('AGE' , fontsize = 20)
plt.xlabel('Age Category' , fontsize = 15)
plt.ylabel('No. of Students' , fontsize = 15)
plt.show()
```

```python
# 3) Bar Plot : A bar plot or bar chart is a graph that represents the category of data with rectangular bars with 
# lengths and heights that is proportional to the values which they represent.

import numpy as np 
import matplotlib.pyplot as plt
from matplotlib import style
style.use('ggplot')

classes = ['Math' , 'Science' , 'History' , 'Geography']
students = [54 , 34 , 45 , 59]
plt.bar(classes , students , width=0.6 , color='m')  # for horizontal bar plot just use plt.barh()
plt.title('Class Count' , fontsize = 20)
plt.xlabel('Class Name' , fontsize = 15)
plt.ylabel('No. of Students' , fontsize = 15)
plt.show()
```

```python
# 4) Scatter Plot : A scatter plot is a diagram where each value in the data set is represented by a dot. 

import numpy as np 
import matplotlib.pyplot as plt
from matplotlib import style
style.use('ggplot')

import pandas as pd
from sklearn.datasets import load_wine    #scikit-learn is basically a ML python library to load datasets 

data = load_wine()
wine_arr = pd.DataFrame(data["data"] , columns = data['feature_names'])
#print(wine_arr)

res1 = wine_arr['magnesium']
res2 = wine_arr['color_intensity']
#print(res1)
#print(res2)
plt.figure(figsize=(10,8))
plt.scatter(res1 , res2 , c='r' ,marker='o',s=110, alpha=0.9)  # alpha for intensity of dots , s=size of dot
plt.title('Components Of Drink' , fontsize = 20)
plt.xlabel('Magnesium' , fontsize = 15)
plt.ylabel('color_intensity' , fontsize = 15)
plt.show()
```

```python
# 5) Box Plot : A box plot which is also known as a whisker plot displays a summary of a set of data containing 
# the minimum, first quartile, median, third quartile, and maximum.

import numpy as np 
import matplotlib.pyplot as plt
from matplotlib import style
style.use('ggplot')

import pandas as pd
from sklearn.datasets import load_wine    #scikit-learn is basically a ML python library to load datasets 

data = load_wine()
wine_arr = pd.DataFrame(data["data"] , columns = data['feature_names'])
res2 = wine_arr['color_intensity']
plt.figure(figsize=(8,6))
plt.boxplot(wine_arr.alcohol , meanline=True , showmeans=True)
plt.title('Components Of Drink' , fontsize = 20)
plt.ylabel('hue' , fontsize = 15)
plt.savefig('Box_Plot' , dpi=100, quality=98 , facecolor='g')
plt.show()
```
