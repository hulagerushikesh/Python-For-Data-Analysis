```python
#                             Data Visualization

#                  @ Seaborn 

# ----->   Is a pythonn library for data visualization build up on the matplotlib python library.
# --> It is mainly used stastical graphs.

# As in past tutorial we learnt matplotlib for visualization , so where it fails and seaborn take the advantages ?
# Seaboorn Vs Matplotlib :
# -----> Seaborn has large no. of in-biult functions so with in a few line of code we can plot the big / dense data easily ,
# and on other side in matplotlib we have to write multiple line code. But we can plot simple plots with the matplotlib as 
# we are more familiar with it and code is understand by anyone and as seaborn have in-biult functions , so we need to 
# remember all of them and its quit time consuming and complex in case of simple plots.
# in short dennse data---->seaborn and simple plots---->matplotlib.

# What kind of graphs we can plot using seaborn ?
# ---> Heatmap and Pair plot( mainly used in ML) And Facet plot , Bar plot , Scatter plot , Line , Distplot / Histogram
# Violin plot, Boxplot 



import matplotlib.pyplot as plt  # as we know seaborn is biiuld up on matplotlib therefore we are gointg to used it.
import seaborn as sns
import numpy as np 
import pandas as pd
from matplotlib import style
style.use('ggplot')

# 1) Line plot: Used for finding relationships between 2 cols of a dataset and we can plot the relation wrt other col
# we take the data same we used for line plot in matplotlib just for camparison part

plt.figure(figsize=(8,6))
days = np.arange(1 , 31)
temp = np.linspace(25 , 40 , 30)
temp1 = np.array([23,34,45,23,22,33,43,44,34,25,27,32,29,39,31, 
                  23,34,46,42,39,31,27,23,26,29,33,37,31,30,40])
data1_df = pd.DataFrame({'days':days , 'temp':temp})     # seaborn uses dataframe for x and y values 
data2_df = pd.DataFrame({'days':days , 'temp1':temp1})
sns.lineplot(x='days', y='temp', data=data1_df,)  # as we can see in plot we didn't label x and y  asix but 
sns.lineplot(x='days', y='temp1', data=data2_df,) # seabron directly use the x and y parameters as asix
plt.show()
```

```python
# Above is kind of syntax intro for to visualize the x and y plot as a datframe 
# its quit similar as matplotlib ,but for datasets with multuple cols its difficcult t use matplotlib 
# So onwards we are performing operations on a dataset and try to visualise it.

# but here you might think that we need to download the dataset read it using pandas , well seaborn provides sample 
# dataset in there library and we can access them by just a line of code as we load wine dataset using scikit learn in numpy tutorial
import matplotlib.pyplot as plt  # as we know seaborn is biiuld up on matplotlib therefore we are gointg to used it.
import seaborn as sns
import numpy as np 
import pandas as pd
#from matplotlib import style
#style.use('ggplot')


# now intead of using style we are going to use sns.set() function
sns.set(style='darkgrid' , )

tips_df = sns.load_dataset('tips')
print(tips_df.head()) 
plt.figure(figsize=(12,8))
# we can see now the cols of dataset . It has tip amount , total bill ,sex , smoker , day of week , time and the size

# Here we are going to plot total bill vs tips as they are the cils of the dataset , you can use any col vs another col
# in order to check the relation between the cols
sns.lineplot(x='total_bill', y='tip', data=tips_df.head(60), hue='sex'    # using hue we can plot catageries its similar like legend
            ,style='sex', palette='hot',                   # and with just a single word we can five the vals in seaborn
            markers=['o','>'], )                    
plt.show()     
plt.savefig('snslineplot')
```

```python
# 2) Histogram : only plot numerical data wrt to x axis only
import matplotlib.pyplot as plt  # as we know seaborn is biiuld up on matplotlib therefore we are gointg to used it.
import seaborn as sns
import numpy as np 
import pandas as pd
from scipy.stats import norm
#from matplotlib import style
#style.use('ggplot')

# now intead of using style we are going to use sns.set() function
sns.set(style='darkgrid' , )

tips_df = sns.load_dataset('tips')
tips_df.head()
plt.figure(figsize=(12,8))
   #bins is the no histograms 
#bin=[1,5,10,15,20,25,30,35,40,45,50,55,60]
sns.distplot(tips_df['total_bill'] ,bins=5 ,fit=norm , kde=True, color='r',label='Total Bill', rug=True,
            hist_kws={'color':'g','edgecolor':'b','linewidth':3,'alpha':0.9},
            kde_kws={'color':'r','linewidth':3,'alpha':0.6},
            rug_kws={'color':'m','linewidth':3,'alpha':0.9},
            fit_kws={'color':'y','linewidth':3,'alpha':0.9})   
sns.distplot(tips_df['size'],fit=norm , kde=False, color='r',label='Total Bill', rug=False)
sns.distplot(tips_df['tip'],fit=norm , kde=False, color='r',label='Total Bill', rug=False)
plt.legend(loc=0)
plt.title('snsHistogram of Total Bill' , fontsize=20)
plt.show()#kernal density estimator(kde) --> line plot onn histogram
plt.savefig('snsHisogram')
```

```python
# 3) Bar Plot : In bar plot we are going to use both x and y axis. we can plot string staset too
import matplotlib.pyplot as plt  # as we know seaborn is biiuld up on matplotlib therefore we are gointg to used it.
import seaborn as sns
import numpy as np 
import pandas as pd
from scipy.stats import norm
#from matplotlib import style
#style.use('ggplot')

# now intead of using style we are going to use sns.set() function
sns.set(style='darkgrid' , )
plt.figure(figsize=(12,8))

tips_df = sns.load_dataset('tips')
tips_df.head()
kwsargs= {'alpha':0.9, 'linestyle':':', 'linewidth':4, 'edgecolor':'k'}
order=['Sun','Mon','Tues','Wed','Thur','Fri','Sat'] # sorting days
ax = sns.barplot(x='day', y='total_bill', data=tips_df, hue='sex', order=order, estimator=np.mean, ci=22,
           palette='magma', saturation=4, errcolor='0.5', errwidth=6, **kwsargs,)
plt.title('sns BarPlot', fontsize=25)
ax.set(xlabel='Days', ylabel='Total Bill',)
plt.legend(loc=0)
plt.show() # by usnind data argument you need not write tips_df. every time
# ci for line size of trhe estimator
plt.savefig("snsBarPlot")
```

```python
# 4) Scatter Plot :

import matplotlib.pyplot as plt  # as we know seaborn is biiuld up on matplotlib therefore we are gointg to used it.
import seaborn as sns
import numpy as np 
import pandas as pd
from scipy.stats import norm
#from matplotlib import style
#style.use('ggplot')

# now intead of using style we are going to use sns.set() function
sns.set(style='darkgrid' , )
plt.figure(figsize=(12,8))

tips_df = sns.load_dataset('tips')
tips_df.head()

sns.scatterplot(x='day',y='total_bill',data=tips_df, hue='sex', style='smoker', alpha=0.9, sizes=(0,15), )
plt.title('sns Scatter Plot', fontsize=25)
plt.xlabel('Days', fontsize=20)
plt.ylabel('Total Bill', fontsize=20)
plt.legend(loc=0)
plt.show()
```

```python
# 5) Heatmap :  Is a graphical representation of 2-D data.  ( feature selection most use in ML)
# a plot a matrix 
# only plot numerical data

import matplotlib.pyplot as plt  # as we know seaborn is biiuld up on matplotlib therefore we are gointg to used it.
import seaborn as sns
import numpy as np 
import pandas as pd
from scipy.stats import norm
#from matplotlib import style
#style.use('ggplot')

# now instead of using plt.style we are going to use sns.set() function
sns.set(style='darkgrid' , )
plt.figure(figsize=(12,8))

eg = pd.read_csv(r'C:\Users\ADMIN\OneDrive\Desktop\Data Analysis\Iris.csv')
eg.head(20)
new_cols = eg.drop(columns=['Id']).set_index('Species')
new_cols.head(10)

annot_kws = {'fontsize':10,'fontstyle':'italic','alpha':0.9,'color':'k','verticalalignment':'center'}
cbar_kws = {'shrink':4,'extend':'min','extendfrac':0.9,'ticks':np.arange(0,20),'drawedges':True}

sns.heatmap(new_cols.tail(30), cmap='vlag',annot=True, annot_kws=annot_kws, linewidth=1, linecolor='w', cbar=False,
           cbar_kws=cbar_kws)
plt.show()
```

```python
# correlation heatmap
# What is correlation : Is a statistical relationship between two vars(x,y)  (ML part - Feature selection)
# simple x proportional to y
plt.figure(figsize=(10,8))
new_cols.corr()
ax=sns.heatmap(new_cols.corr(), annot=True, linewidth=2)
ax.tick_params(size=3, color='w', labelsize=12,labelcolor='r')
plt.title('snsCorrHeatmap',fontsize=20)
plt.show()
```
