```python


import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import style
style.use('ggplot')

Iris = pd.read_csv(r'C:\Users\ADMIN\OneDrive\Desktop\Data Analysis\Iris.csv')
print(Iris)

Iris.plot(kind='box', subplots=True, layout=(3,2), figsize=(12,16));
Iris.hist(figsize=(12,16) , color='g' , rwidth = 0.8 , grid='r' )
plt.grid(color ='r' , linestyle = '-' , linewidth = 1)
Iris.corr()
sns.heatmap(Iris.corr(), annot=True, cmap ='Wistia');
sns.pairplot(Iris , kind='scatter' , diag_kind='hist' , height=3)
plt.show()
```

```python
# Enhacing Data visualizations using plotnine and ggplot :

# 1) Facets : Facets are used to plot subsets of data. it allows an individual plot for groups of data in the same image.
#For example, let’s consider the tips dataset that contains information about people who probably had food at a
#restaurant and whether or not they left a tip, their age, gender and so on. Lets have a look at it.
#Now let’s suppose we want to plot about what was the total bill according to the gender and on each day. In such
#cases facets can be very useful, let’s see how.



from plotnine import ggplot, aes, facet_grid, labs, geom_col
import numpy as np 
import matplotlib.pyplot as plt
from matplotlib import style
style.use('ggplot')
import pandas as pd
Iris = pd.read_csv(r'C:\Users\ADMIN\OneDrive\Desktop\Data Analysis\Iris.csv')
#print(Iris)



(
ggplot(Iris) + facet_grid(facets="~Species") + aes(x="SepalLengthCm", y="PetalLengthCm" ) + geom_col()
)
```

```python
# 2) Statistical transformations :
from plotnine import ggplot, aes, facet_grid, labs, geom_histogram
import numpy as np 
import matplotlib.pyplot as plt
from matplotlib import style
style.use('ggplot')
import pandas as pd
Iris = pd.read_csv(r'C:\Users\ADMIN\OneDrive\Desktop\Data Analysis\Iris.csv')
#print(Iris)



(
ggplot(Iris) + facet_grid(facets="~Species") + aes(x="SepalLengthCm" ) + geom_histogram(bins=10)
)
```

```python
# 3) Coordinates :
from plotnine import ggplot, aes, facet_grid, labs,geom_histogram, coord_flip
import numpy as np 
import matplotlib.pyplot as plt
from matplotlib import style
style.use('ggplot')
import pandas as pd
Iris = pd.read_csv(r'C:\Users\ADMIN\OneDrive\Desktop\Data Analysis\Iris.csv')
#print(Iris)



(
ggplot(Iris) + facet_grid(facets="~Species") + aes(x="SepalLengthCm" ) + geom_histogram(bins=10) + coord_flip()
)
```
